```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('coins.jpg',0)


img = cv2.medianBlur(img,5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
edges = cv2.Canny(cimg,200,250)
circles = cv2.HoughCircles(edges,cv2.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))    # PERFECT DETECTION WITH LINES
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

# Draw a diagonal vertical line with thickness of 5 px
img = cv2.line(cimg,(60,0),(60,500),(255,0,0),5)
img = cv2.line(img,(150,0),(150,500),(255,0,0),5)
img = cv2.line(img,(300,0),(300,500),(255,0,0),5)
img = cv2.line(img,(480,0),(480,500),(255,0,0),5)

plt.imshow(img)
plt.show()
cv2.imshow('detected circles',img)
cv2.waitKey(0)
cv2.destroyAllWindows()




```
